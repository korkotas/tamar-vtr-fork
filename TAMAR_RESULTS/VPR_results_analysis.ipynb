```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/results_chan_width_100_iters_50_frequency_1.xlsx')
```

```python
df.drop(df.columns[0], axis = 1, inplace = True)
new_row_names = ['Iterations', 'Heap Pushes', 'Heap Pops', 'Iterations w/o', 'Heap Pushes w/o', "Heap Pops w/o"]
df.index = new_row_names
```

```python
df
```

```python
# Calculate the percentage change
iterations_change = ((df.loc['Iterations'] - df.loc['Iterations w/o']) * 100 / df.loc['Iterations w/o'])
heap_pushes_change = ((df.loc['Heap Pushes'] - df.loc['Heap Pushes w/o']) * 100 / df.loc['Heap Pushes w/o'])
heap_pops_change = ((df.loc['Heap Pops'] - df.loc['Heap Pops w/o']) * 100 / df.loc['Heap Pops w/o'])

# Add the percentage changes as new rows
df.loc['Iterations Change (%)'] = iterations_change
df.loc['Heap Pushes Change (%)'] = heap_pushes_change
df.loc['Heap Pops Change (%)'] = heap_pops_change
```

```python
df
```

```python
# Select the rows for the bar graph
heap_pushes = df.loc['Heap Pushes']
heap_pushes_wo = df.loc['Heap Pushes w/o']

# Create the bar graph
plt.figure(figsize=(20, 6))
# Parameters for bars
bar_width = 0.35
index = range(len(heap_pushes))  # Use a range based on the number of elements

# Plotting bars
plt.bar([i - bar_width/2 for i in index], heap_pushes, bar_width, label='Heap Pushes in Dynamic')
plt.bar([i + bar_width/2 for i in index], heap_pushes_wo, bar_width, label='Heap Pushes in Map')

# Labels and title
plt.xlabel('Columns')
plt.ylabel('Heap Pushes')
plt.title('Dynamic Lookahead vs Map Lookahead')
plt.xticks(index, heap_pushes.index)  # Set the x-ticks to match the index of your DataFrame
plt.legend()

# Show plot
plt.show()
```

```python
# Select the rows for the bar graph
heap_pops = df.loc['Heap Pops']
heap_pops_wo = df.loc['Heap Pops w/o']

# Create the bar graph
plt.figure(figsize=(20, 6))
# Parameters for bars
bar_width = 0.35
index = range(len(heap_pops))  # Use a range based on the number of elements

# Plotting bars
plt.bar([i - bar_width/2 for i in index], heap_pops, bar_width, label='Heap Pops in Dynamic')
plt.bar([i + bar_width/2 for i in index], heap_pops_wo, bar_width, label='Heap Pops in Map')

# Labels and title
plt.xlabel('Columns')
plt.ylabel('Heap Pops')
plt.title('Dynamic Lookahead vs Map Lookahead')
plt.xticks(index, heap_pops.index)  # Set the x-ticks to match the index of your DataFrame
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
from scipy.stats import gmean

def calculate_geometric_mean(percentage_changes):
    return np.mean(percentage_changes)
    # Convert percentage changes to multiplicative factors
    factors = 1 + (percentage_changes / 100.0)

    # Calculate the geometric mean of the factors
    geo_mean_factor = gmean(factors)

    # Convert the geometric mean factor back to a percentage change
    geo_mean_percentage = (geo_mean_factor - 1) * 100

    return geo_mean_percentage

# Extract the relevant rows for percentage changes
iterations_change = df.loc['Iterations Change (%)']
heap_pushes_change = df.loc['Heap Pushes Change (%)']
heap_pops_change = df.loc['Heap Pops Change (%)']

# Calculate the geometric mean percentage changes
avg_iterations_change = calculate_geometric_mean(iterations_change)
avg_heap_pushes_change = calculate_geometric_mean(heap_pushes_change)
avg_heap_pops_change = calculate_geometric_mean(heap_pops_change)

print(f'Average Percentage Change in Iterations: {avg_iterations_change:.2f}%')
print(f'Average Percentage Change in Heap Pushes: {avg_heap_pushes_change:.2f}%')
print(f'Average Percentage Change in Heap Pops: {avg_heap_pops_change:.2f}%')
```
